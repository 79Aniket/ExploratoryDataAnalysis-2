## Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

## Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
import seaborn as sns
import pandas as pd

df = pd.read_excel('flight_price.xlsx')

# Create a box plot of price of flights by airline
sns.boxplot(
    x = 'Airline',
    y = 'Price',
    showmeans=True,
    data=df
)

# Rotate x labels to prevent overlapping
plt.xticks(rotation=50)
plt.show()
```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd

# Load the Google Playstore dataset
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Get the number of installs for each app
number_of_installs = df['Installs'].value_counts()

# Get the top 10 most popular apps
top_10_apps = number_of_installs.head(10)

# Create a frequency table of the top 10 most popular apps
frequency_table = pd.DataFrame(top_10_apps).reset_index()
frequency_table.columns = ['App', 'Number of Installs']

print(frequency_table)
```
